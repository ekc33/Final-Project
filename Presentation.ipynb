```python

df1 = pd.read_csv("contract-rent.csv")

r = df1[df1['Estimate; Total:'] < 500]

r = r.dropna()

r.plot.bar(x = 'Neighborhood', y = 'Estimate; Total:').set_title('Neighborhoods with rent less than $500')
```

```python
pp = pd.read_csv("median_dead.csv")
pp = pp[pp["TOTAL MD AGE AT DEATH"]>80]
pp.plot.bar(x = 'NEIGHBORHOOD', y = 'TOTAL MD AGE AT DEATH' )
```

This shows the median age of death of neighborhoods in pittsburgh. The median age is not significantly different, but it can be a strong indicator of which neighborhood has high life expectancy. At first glance it looks like Squirrel Hill has the highest.
